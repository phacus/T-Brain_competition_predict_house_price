```python
train = pd.read_csv('dataset-0510/train.csv')
test  = pd.read_csv('dataset-0510/test.csv')
data = pd.concat([train, test], axis=0)
data['house_price'] = data['total_price'] - data['parking_price']
data['house_area']  = data['land_area'] - data['parking_area']
```

```python
qq = data[(data['parking_price'].notna()) & (data['total_price'].notna())]
pp = data[(data['parking_price'].isna()) & (data['total_price'].notna())]
cols = ['house_price', 'house_area', 'city', 'building_material', 'building_type', 'building_use', 'parking_way', 'parking_price', 'total_floor', 'txn_floor']
```

```python
#categorical feature to one-hot
def one_hot(train, test, categorical_features):
    for i in categorical_features:
        train = train.join(pd.get_dummies(data[i], prefix = i))
        test = test.join(pd.get_dummies(data[i], prefix = i))
        train.drop(i, axis = 1, inplace =True)
        test.drop(i, axis = 1, inplace =True)
        
    return train, test
```
